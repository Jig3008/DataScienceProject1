# **## Step 1: Load the data**

```python
#Welcome to the sample project of Python Pandas and Data Wrangling
```

```python
#import the pandas library
import pandas as pd
#get the shares data of reliance in the sharesData DataFrame
sharesData = pd.read_csv('relianceSharesData.csv')
sharesData.head(10)
```

# **## Step 2: After the intial data is loaded. We would work on cleaning the data:[Data Wrangling]**
1. Assign the right data types to columns
2. Drop the columns which are not needed
3. Remove the leading or trailing spaces present

```python
#Step2 : Identify which columns have the data that are needed and the data types of each column
#Also check if the column name themselves have any leading or trailing spaces

sharesData.columns
```

```python
#As we see from the above there are spaces present - we need to correct them
sharesData.columns = sharesData.columns.str.strip()

sharesData.columns
```

```python
#Check the data types of each column present

sharesData.info()
```

```python
#after checking the vaue its apparent that the columns are not tagged to the right data types - lets correct them
#We would also need to correct them we need to remove the , from the decimal values

floatColumns = ['OPEN','HIGH','LOW','PREV. CLOSE','ltp','close','vwap','52W H','52W L','VALUE']

for col in floatColumns:
  sharesData[col] = sharesData[col].str.replace(pat=',',repl='',regex=False).astype('float64')

sharesData.info()

intColumns = ['VOLUME','No of trades']

for col in intColumns:
  sharesData[col] = sharesData[col].str.replace(pat=',',repl='',regex=False).astype('int64')

sharesData.info()
```

```python
#Lets check the series column - as it seems categorical data - what data is truly present

sharesData['series'].value_counts()
```

```python
#We know this data is for equity so we would remove the column, we also do not need the LTP, Close, Vwap columns
columns_to_drop = ['ltp','series','close','vwap']
sharesData = sharesData.drop(
    labels=columns_to_drop,
    axis =1
)

sharesData.head()
```

```python
#While we are at it let us work on the column names too - as they are not in symmetry

sharesData.columns = sharesData.columns.str.title()
sharesData.info()
```

# **## Step 3: Data Analysis**
1. What day was the highest number of trades done?
2. Check the difference between the highest and lowest trading value for each day?
3. Check which day had the highest difference between the opening and closing?
4. Check for a specific day - what was the opening and closing values?
5. Find the percentage changes between the opening and closing values?

```python
highestTrade = sharesData.sort_values(by='No Of Trades',ascending=False)
highestTrade.head(10)
```

```python
sharesData['Difference Opening/Closing'] = sharesData['High'] - sharesData['Low']
sharesData['Difference Percentage'] = (sharesData['Difference Opening/Closing']/sharesData['Low'])*100
sharesData.head()
```

```python
#lets analyze the statistical values for the column
sharesData['Difference Percentage'].describe()
```

```python
#looking at the above it seems that there are instances where the max deviation is more than 10

shares_Deviation_gt10 = sharesData[sharesData['Difference Percentage'] > 10]
shares_Deviation_gt10.head()
```
